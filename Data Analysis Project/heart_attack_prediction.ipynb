# Introduction
Predicting the likelihood of a heart attack is an important task in healthcare, as early detection and intervention can greatly improve the outcome for patients. There are many factors that can increase the risk of a heart attack, including age, genetics, lifestyle, and medical history. Data analysis techniques can be used to identify patterns in data that may be indicative of an increased risk of heart attack, and to develop models that can predict the likelihood of a heart attack in an individual.

## Building the Classification Model

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier() 
```

```python
# independent variables (predictors)
X_heart = heart[['Chlstrl', 'Age', 'Max HR', 'ECG','Sex', 'CP', 'Bld Prssr', 'Bld Sgr', 'ECG',
       'Ex Angina', 'Vessels']]
y_heart = heart['HA Risk']
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_heart, y_heart, random_state=0)
```

```python
ytest.groupby(ytest).size()
```

```python
tree.fit(Xtrain, ytrain);
ypred = tree.predict(Xtest)
from sklearn.metrics import accuracy_score
acc = accuracy_score(ytest, ypred)
print("Accuracy: %0.2f%%" % (100 * acc))
```

## Analyzing the Results

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(ytest, ypred, labels=tree.classes_)
cm
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(2.6, 2.6))
ax = sns.heatmap(cm, vmin=0, vmax=0, cmap='gray_r', annot=True,
linewidths=0.5, linecolor='k', cbar=False,
xticklabels=tree.classes_, yticklabels=tree.classes_);
ax.set(title='Confusion matrix', xlabel='Predicted label', ylabel='True label')
# re-enable outer spines
sns.despine(left=False, right=False, top=False, bottom=False)
```
